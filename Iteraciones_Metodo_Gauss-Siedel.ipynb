```
#Ejercicio 1
def gs11(y, z):
    return (-.222 - (.222*y) + (.111*z))
def gs12(x, z):
    return ((.375) - (.875*x) - (.625*z))
def gs13(x, y):
    return ((-.600) + (.300*x) + (.400*y))

#Ejercicio 2
def gs21(y):
    return (2.4 - (.4*y))
def gs22(x):
    return ((.8) + (.1*x))

#Ejercicio 3
def gs31(y):
    return (.5 - (.125*y))
def gs32(x):
    return ((.6) - (.4*x))
def gs33(x):
    return ((.75) - (.25*x))

#Ejercicio 4
def gs41(y, z):
    return (3.667 - (.333*y) - (.167*z))
def gs42(x, z):
    return ((2.5) + (.125*x) - (.25*z))
def gs43(x, y):
    return ((3.833) - (.167*x) + (.167*y))
```

```
import matplotlib.pyplot as plt
import math

def GaussSeidel():
  print("Metodo Gauss-Seidel")
  print("\n Ejercicio 1 \n")
  Gauss_Seidel_Ejercicio1()
  print("\n Ejercicio 2 \n")
  Gauss_Seidel_Ejercicio2()
  print("\n Ejercicio 3 \n")
  Gauss_Seidel_Ejercicio3()
  print("\n Ejercicio 4 \n")
  Gauss_Seidel_Ejercicio4()

#Ejercicio 1
def Gauss_Seidel_Ejercicio1():
  xDat=[]
  xDat.append(0)
  yDat=[]
  yDat.append(0)
  zDat=[]
  zDat.append(0)
  cont = 0
  x = 0
  y = 0
  z = 0
  while(cont < 4):
    xd = gs11(y, z)
    xDat.append(xd)
    x = xd
    yd = gs12(x , z)
    yDat.append(yd)
    y = yd
    zd = gs13(x, y)
    zDat.append(zd)
    z = zd
    cont += 1
  plt.plot(xDat, label = "x_Dato")
  plt.plot(yDat, label = "y_Dato")
  plt.plot(zDat, label = "z_Dato")
  plt.legend()
  plt.show()
  print("Iteraciones: \n")
  print("x : " + str(xDat))
  print("y : " + str(yDat))
  print("z : " + str(zDat))

#Ejercicio 2
def Gauss_Seidel_Ejercicio2():
  xDat=[]
  xDat.append(0)
  yDat=[]
  yDat.append(0)
  cont = 0
  x = 0
  y = 0
  while(cont < 4):
    xd = gs21(y)
    xDat.append(xd)
    x = xd
    yd = gs22(x)
    yDat.append(yd)
    y = yd
    cont += 1
  plt.plot(xDat, label = "x_Dato")
  plt.plot(yDat, label = "y_Dato")
  plt.legend()
  plt.show()
  print("Iteraciones: \n")
  print("x : " + str(xDat))
  print("y : " + str(yDat))

#Ejercicio 3
def Gauss_Seidel_Ejercicio3():
  xDat=[]
  xDat.append(0)
  yDat=[]
  yDat.append(0)
  zDat=[]
  zDat.append(0)
  cont = 0
  x = 0
  y = 0
  z = 0
  while(cont < 4):
    xd = gs31(y)
    xDat.append(xd)
    x = xd
    yd = gs32(x)
    yDat.append(yd)
    y = yd
    zd = gs33(x)
    zDat.append(zd)
    z = zd
    cont += 1
  plt.plot(xDat, label = "x_Dato")
  plt.plot(yDat, label = "y_Dato")
  plt.plot(zDat, label = "z_Dato")
  plt.legend()
  plt.show()
  print("Iteraciones: \n")
  print("x : " + str(xDat))
  print("y : " + str(yDat))
  print("z : " + str(zDat))

#Ejercicio 4
def Gauss_Seidel_Ejercicio4():
  xDat=[]
  xDat.append(0)
  yDat=[]
  yDat.append(0)
  zDat=[]
  zDat.append(0)
  cont = 0
  x = 0
  y = 0
  z = 0
  while(cont < 4):
    xd = gs41(y, z)
    xDat.append(xd)
    x = xd
    yd = gs42(x , z)
    yDat.append(yd)
    y = yd
    zd = gs43(x, y)
    zDat.append(zd)
    z = zd
    cont += 1
  plt.plot(xDat, label = "x_Dato")
  plt.plot(yDat, label = "y_Dato")
  plt.plot(zDat, label = "z_Dato")
  plt.legend()
  plt.show()
  print("Iteraciones: \n")
  print("x : " + str(xDat))
  print("y : " + str(yDat))
  print("z : " + str(zDat))

GaussSeidel()
```
