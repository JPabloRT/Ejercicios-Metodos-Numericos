```
#Ejercicio 1
def j11(y, z):
    return (-.222 - (.222*y) + (.111*z))
def j12(x, z):
    return ((.375) - (.875*x) - (.625*z))
def j13(x, y):
    return ((-.600) + (.300*x) + (.400*y))

#Ejercicio 2
def j21(y):
    return (2.4 - (.4*y))
def j22(x):
    return ((.8) + (.1*x))

#Ejercicio 3
def j31(y):
    return (.5 - (.125*y))
def j32(x):
    return ((.6) - (.4*x))
def j33(x):
    return ((.75) - (.25*x))

#Ejercicio 4
def j41(y, z):
    return (3.667 - (.333*y) - (.167*z))
def j42(x, z):
    return ((2.5) + (.125*x) - (.25*z))
def j43(x, y):
    return ((3.833) - (.167*x) + (.167*y))
```
